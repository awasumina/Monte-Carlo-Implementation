```python
import numpy as np
import matplotlib.pyplot as plt

radius = 4
N = 1000

# Generate random points
x_random = np.random.uniform(0, radius, N)
y_random = np.random.uniform(0, radius, N)

# Determine which points are inside the quarter circle
inside_circle = (x_random**2 + y_random**2) <= radius**2

# Number of points inside the circle
n = np.sum(inside_circle)

# Area of the quarter circle
area_quarter_circle = (n / N) * (radius * radius)

# Estimate of π
pi_estimate = (area_quarter_circle * 4) / (radius * radius)

# Display results
print(f"Estimated π={pi_estimate}")
print(f"Analytical π={np.pi}")
print(f"Error = {abs(np.pi - pi_estimate)}")
print(f"Error % = {abs(np.pi - pi_estimate) / np.pi * 100:.2f}%")

# Plot
theta = np.linspace(0, np.pi / 2, 1000)
x_circle = radius * np.cos(theta)
y_circle = radius * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x_circle, y_circle, color='blue', label='Quarter circle')
plt.scatter(x_random[inside_circle], y_random[inside_circle], color='green', s=1, label='Points inside circle')
plt.scatter(x_random[~inside_circle], y_random[~inside_circle], color='red', s=1, label='Points outside circle')
plt.xlim(0, radius)
plt.ylim(0, radius)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Monte Carlo Method to Estimate π')
plt.legend()
plt.grid(True)
plt.show() 
```
